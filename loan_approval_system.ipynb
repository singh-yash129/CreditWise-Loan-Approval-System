# Handling Missing Data

```python
numerical_cols = loan_df.select_dtypes(include=["number"]).columns
```

```python
numerical_cols
```

```python
categorical_cols = loan_df.select_dtypes(include=["object"]).columns
```

```python
categorical_cols
```

# fill the missing data use imputer

```python
num_imp = SimpleImputer(strategy = "mean")
loan_df[numerical_cols] = num_imp.fit_transform(loan_df[numerical_cols])
```

```python
cgt_imp = SimpleImputer(strategy = "most_frequent")
loan_df[categorical_cols] = cgt_imp.fit_transform(loan_df[categorical_cols])
```
